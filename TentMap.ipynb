```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Tend_map(x,L):
    return L*np.minimum(x,1-x)
```

```python
x=[0.001]
R=2
x_total = np.linspace(0,1,1000)
for i in range(100):
    x.append(Tend_map(x[-1],R))
    plt.step(x,x,alpha=1,color="rebeccapurple",linewidth=0.8)
    plt.title(r"$\lambda=${}, iteration number = {}".format(round(R,3),i+1))
    plt.plot(x_total,Tend_map(x_total,R),color="navy")
    plt.plot(x_total,x_total,color="orange")
    plt.xlim(0,0.6)
    plt.ylim(0,0.6)
    plt.savefig(str(i).zfill(4)+".png")
    plt.show()
```

```python
plt.figure(figsize=(10,10))
for L in np.linspace(1,4,1000):
    x=np.linspace(0.0001,1,10000)
    for i in range(100):
        x=Tend_map(x,L)
    plt.plot([L]*len(x),x,".",markersize=0.01,color="k")
    plt.ylim(0,1)
    plt.xlim(1.0,2)
plt.savefig("BTM.png")
plt.show()
```
